```python
import math

# Pedir al usuario los coeficientes de la ecuación cuadrática
a = float(input("Ingresa el valor de a: "))
b = float(input("Ingresa el valor de b: "))
c = float(input("Ingresa el valor de c: "))

# Calcular el discriminante
discriminante = b**2 - 4*a*c

# Calcular las dos soluciones usando la fórmula general
raiz1 = (-b + math.sqrt(discriminante)) / (2*a)
raiz2 = (-b - math.sqrt(discriminante)) / (2*a)

# Mostrar resultados
print(f"Las soluciones de la ecuación son: {raiz1} y {raiz2}")
```

```python
# Programa para sumar los primeros N números naturales sin usar condicionales

# Solicitar al usuario el número hasta donde desea sumar
n = int(input("¿Hasta qué número natural quieres sumar? "))

# Calcular la suma con la fórmula N(N + 1) / 2
suma = n * (n + 1) // 2  # División entera para obtener un número entero

# Mostrar el resultado
print(f"La suma de los primeros {n} números naturales es: {suma}")
```

```python
# Condicionales


#¿para que van a servir? PARA QUE SE EJECUTE UNA ACCION O SE EJECUTE OTRA O NO SE
#EJECUTE NADA

numero = int(input("Digite un numero: ")) #VERIFIACAMOS SI ES POSITIVO

if numero > 0: #if= SI, SI SE CUMPLE VAMOS A DETERMINAR ACCIONES, NOTAAA EN OTROS LENGUAJES USAMOS {} PARA SABER QUE ESTAN DENTRO DEL IF
    print("El numero es positivo") #AQUI VAN DENTRO DE LA MISMA IDENTACION O ES DECIR ESTEN ALINEADOS ESTARAN DENTRO DEL IF
elif numero == 0:
    print("El numero es cero")
else:      #SI NO, CASO CONTRARIO
    print("El numero es negativo")
```
